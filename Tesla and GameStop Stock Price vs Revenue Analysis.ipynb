```python
# Importing Libraries
import pandas as pd
import yfinance as yf
import matplotlib.pyplot as plt
import seaborn as sns
print("All modules imported successfully!")
```

```python
# Fetch historical data for Tesla  
tesla = yf.Ticker("TSLA")


tesla_prices = tesla.history(period="5y")


# Display the first few rows
tesla_prices.head()
```

```python
# Fetch historical data for  GameStop
gamestop = yf.Ticker("GME")


gamestop_prices = gamestop.history(period="5y")

# Display the first few rows
gamestop_prices.head()
```

```python
# Create revenue data manually
tesla_revenue = pd.DataFrame({
    "Date": ["2023-03-31", "2023-06-30", "2023-09-30"],
    "Revenue": [23000, 25000, 27000]
})

gamestop_revenue = pd.DataFrame({
    "Date": ["2023-03-31", "2023-06-30", "2023-09-30"],
    "Revenue": [1500, 1700, 1800]
})

# Convert dates to datetime format
tesla_revenue["Date"] = pd.to_datetime(tesla_revenue["Date"])
gamestop_revenue["Date"] = pd.to_datetime(gamestop_revenue["Date"])

# Display Tesla revenue
tesla_revenue
```

```python
# Create revenue data manually
 
gamestop_revenue = pd.DataFrame({
    "Date": ["2023-03-31", "2023-06-30", "2023-09-30"],
    "Revenue": [1500, 1700, 1800]
})

# Convert dates to datetime format
gamestop_revenue["Date"] = pd.to_datetime(gamestop_revenue["Date"])

# Display Tesla revenue
gamestop_revenue
```

```python
plt.figure(figsize=(12, 6))

plt.plot(tesla_prices.index, tesla_prices['Close'], label='Tesla Stock Price', color='blue')
 
plt.title('Stock Prices Over Time')
plt.xlabel('Date')
plt.ylabel('Price (USD)')
plt.legend()
plt.grid()
plt.show()
```

```python
plt.figure(figsize=(12, 6))

plt.plot(gamestop_prices.index, gamestop_prices['Close'], label='GameStop Stock Price', color='red')

plt.title('Stock Prices Over Time')
plt.xlabel('Date')
plt.ylabel('Price (USD)')
plt.legend()
plt.grid()
plt.show()
```

```python
plt.figure(figsize=(10, 5))
x = range(len(tesla_revenue))

plt.bar(x, tesla_revenue["Revenue"], width=0.4, label='Tesla', align='center', color='blue')
plt.bar([p + 0.4 for p in x], gamestop_revenue["Revenue"], width=0.4, label='GameStop', align='center', color='red')

plt.title('Revenue Comparison')
plt.xticks([p + 0.2 for p in x], tesla_revenue["Date"].dt.strftime('%Y-%m'))
plt.ylabel('Revenue (in millions)')
plt.xlabel('Quarter')
plt.legend()
plt.show()
```
